```python
def to_tensor(x, dtype=np.float32):
    if isinstance(x, torch.Tensor):
        return x
    x = np.asarray(x, dtype=dtype)
    x = torch.from_numpy(x)
    return x
```

```python
class BackbonedQ(nn.Module):
    def __init__(self, action_dim : int, hidden_dim : int, net_output_dim : int, activation : nn.Module = nn.LeakyReLU):
        super().__init__()
        self.Q = nn.Sequential(
            nn.Linear(net_output_dim + action_dim, hidden_dim),
            nn.LeakyReLU(),
            nn.Linear(hidden_dim, 1)
        )

    def forward(self, net_pass : torch.Tensor, action : int):
        action_one_hot = F.one_hot(action, num_classes=self.Q[0].in_features - net_pass.shape[1]).float()
        return self.Q(torch.cat([net_pass, action_one_hot], dim=1))
```

```python
class TrajectoryBuffer:
    def __init__(self, gamma: float):
        self.gamma = gamma
        self.trajectories: List[List[Tuple[torch.Tensor, int, torch.Tensor, float]]] = []
        self.current_trajectory: List[Tuple[torch.Tensor, int, torch.Tensor, float]] = []

    def add_step(self, state: torch.Tensor, action: int, log_prob: torch.Tensor, reward: float):
        self.current_trajectory.append((state, action, log_prob, reward))

    def end_episode(self):
        if self.current_trajectory:
            self.trajectories.append(self.current_trajectory)
            self.current_trajectory = []

    def clear(self):
        self.trajectories = []
        self.current_trajectory = []

    def __len__(self):
        return len(self.trajectories)

    def compute_returns(self):
        for traj in self.trajectories:
            rewards = [step[3] for step in traj]
            R = 0.0
            returns = []
            for r in reversed(rewards):
                R = r + self.gamma * R
                returns.append(R)
            returns = torch.tensor(returns[::-1], dtype=torch.float32)
            for (s, a, log_prob, _), ret in zip(traj, returns):
                yield s, a, log_prob, ret

    def compute_advantages_rloo(self):
        """Yields (state, action, log_prob, advantage) for RLOO baseline"""
        if not self.trajectories:
            return
        all_returns: List[torch.Tensor] = []
        for traj in self.trajectories:
            rewards = [step[3] for step in traj]
            R = 0.0
            ret = []
            for r in reversed(rewards):
                R = r + self.gamma * R
                ret.append(R)
            all_returns.append(torch.tensor(ret[::-1], dtype=torch.float32))
        max_T = max(len(rets) for rets in all_returns) if all_returns else 0
        sum_by_t = [0.0] * max_T
        count_by_t = [0] * max_T

        for rets in all_returns:
            for t, gt in enumerate(rets):
                sum_by_t[t] += float(gt)
                count_by_t[t] += 1
        for traj, rets in zip(self.trajectories, all_returns):
            for t, (s, a, log_prob, _) in enumerate(traj):
                if count_by_t[t] > 1:
                    baseline = (sum_by_t[t] - float(rets[t])) / (count_by_t[t] - 1)
                else:
                    baseline = 0.0
                adv = float(rets[t]) - baseline
                yield s, a, log_prob, torch.tensor(adv, dtype=torch.float32)

    def compute_advantages_value(self, value_net: nn.Module):
        """Yields (state, action, log_prob, advantage) using value function baseline"""
        for traj in self.trajectories:
            rewards = [step[3] for step in traj]
            states = [step[0] for step in traj]
            log_probs = [step[2] for step in traj]
            actions = [step[1] for step in traj]
            R = 0.0
            returns = []
            for r in reversed(rewards):
                R = r + self.gamma * R
                returns.append(R)
            returns = torch.tensor(returns[::-1], dtype=torch.float32)
            states_tensor = torch.stack(states)
            with torch.no_grad():
                values = value_net(states_tensor).squeeze(-1)
            advantages = returns - values

            for s, a, log_prob, adv in zip(states, actions, log_probs, advantages):
                yield s, a, log_prob, adv

    def compute_advantages_historical(self, historical_baseline: np.ndarray, decay: float = 0.96):
        """Yields (state, action, log_prob, advantage) using historical reward baseline"""
        for traj in self.trajectories:
            for t, (s, a, log_prob, r) in enumerate(traj):
                if t < len(historical_baseline):
                    baseline = historical_baseline[t]
                else:
                    baseline = 0.0
                adv = r - baseline
                yield s, a, log_prob, torch.tensor(adv, dtype=torch.float32)
        for traj in self.trajectories:

            rewards = [step[3] for step in traj]
            R = 0.0
            returns = []
            for r in reversed(rewards):
                R = r + self.gamma * R
                returns.append(R)
            returns = torch.tensor(returns[::-1], dtype=torch.float32)

            for t, r in enumerate(returns):
                if t < len(historical_baseline):
                    historical_baseline[t] = decay * historical_baseline[t] + (1 - decay) * r
```

```python
class PolicyNetwork(nn.Module):
    def __init__(self, input_dim: int, output_dim: int, hidden_dims: List[int]):
        super().__init__()
        layers = []
        prev_dim = input_dim
        for h_dim in hidden_dims:
            layers.append(nn.Linear(prev_dim, h_dim))
            layers.append(nn.LeakyReLU())
            prev_dim = h_dim
        layers.append(nn.Linear(prev_dim, output_dim))
        self.policy = nn.Sequential(*layers)

    def forward(self, state: torch.Tensor):
        logits = self.policy(state)
        dist = Categorical(logits=logits)
        action = dist.sample()
        log_prob = dist.log_prob(action)
        entropy = dist.entropy()
        return action, log_prob, entropy
    
    def logits(self, state: torch.Tensor):
        return self.policy(state)

    def evaluate(self, state: torch.Tensor) -> int:
        """Deterministic action"""
        with torch.no_grad():
            logits = self.policy(to_tensor(state).float())
            if logits.dim() == 2:
                logits = logits.squeeze(0)
            return int(torch.argmax(logits).item())
```

```python
class ValueNetwork(nn.Module):
    def __init__(self, input_dim: int, hidden_dims: List[int]):
        super().__init__()
        layers = []
        prev_dim = input_dim
        for h_dim in hidden_dims:
            layers.append(nn.Linear(prev_dim, h_dim))
            layers.append(nn.LeakyReLU())
            prev_dim = h_dim
        layers.append(nn.Linear(prev_dim, 1))
        self.net = nn.Sequential(*layers)

    def forward(self, state: torch.Tensor) -> torch.Tensor:
        return self.net(state)
```

```python
class PolicyAgent:
    def __init__(
        self,
        state_dim: int,
        action_dim: int,
        hidden_dims: List[int],
        lr: float = 3e-4,
        gamma: float = 0.99,
        baseline: str = "none",
        value_hidden_dims: List[int] = None,
        entropy_coef: float = 0.0,
        historical_rolling_size: int = 1000,
    ):
        self.gamma = gamma
        self.baseline = baseline.lower()
        self.entropy_coef = entropy_coef
        self.policy = PolicyNetwork(state_dim, action_dim, hidden_dims)
        self.opt = torch.optim.Adam(self.policy.parameters(), lr=lr)
        self.buffer = TrajectoryBuffer(gamma)
        self.use_value_baseline = self.baseline == "value"
        if self.use_value_baseline:
            value_hidden_dims = value_hidden_dims or hidden_dims
            self.value_net = ValueNetwork(state_dim, value_hidden_dims)
            self.value_opt = torch.optim.Adam(self.value_net.parameters(), lr=lr)
        else:
            self.value_net = None
            self.value_opt = None
        self.use_historical_baseline = self.baseline == "historical"
        if self.use_historical_baseline:
            self.historical_baseline = np.zeros(historical_rolling_size)
        else:
            self.historical_baseline = None


        self.use_rloo_baseline = self.baseline == "rloo"

        self._last_log_prob: Optional[torch.Tensor] = None

    def act(self, state: np.ndarray) -> Tuple[int, torch.Tensor, torch.Tensor]:
        s_t = to_tensor(state).float()
        action, log_prob, entropy = self.policy(s_t)
        return int(action.item()), log_prob, entropy

    def store_step(self, state: np.ndarray, action: int, reward: float, log_prob : torch.Tensor, done: bool):
        s_t = to_tensor(state).float()
        self.buffer.add_step(s_t, action, log_prob, reward)
        if done:
            self.buffer.end_episode()

    def update(self, entropy_coef: Optional[float] = None) -> Tuple[float, float] | None:
        if len(self.buffer) == 0:
            return None

        if entropy_coef is not None:
            self.entropy_coef = entropy_coef

        total_loss = 0.0
        value_loss = torch.tensor(0.0)

        if self.use_value_baseline and self.value_opt is not None:
            value_loss = self._update_value()
            total_loss += value_loss
            self.value_opt.zero_grad()
            value_loss.backward()
            self.value_opt.step()
        self.opt.zero_grad()
        policy_loss = self._update_policy()
        total_loss += policy_loss
        policy_loss.backward()
        self.opt.step()

        self.buffer.clear()
        return policy_loss.detach().item(), value_loss.detach().item()

    def _update_policy(self) -> torch.Tensor:
        log_probs: List[torch.Tensor] = []
        advantages: List[torch.Tensor] = []
        entropies: List[torch.Tensor] = []
        if self.use_value_baseline:
            states_train = self.buffer.compute_advantages_value(self.value_net)
        elif self.use_rloo_baseline:
            states_train = self.buffer.compute_advantages_rloo()
        elif self.use_historical_baseline:
            states_train = self.buffer.compute_advantages_historical(self.historical_baseline)
        else:
            states_train = self.buffer.compute_returns()

        for s, _, log_prob, adv in states_train:
            log_probs.append(log_prob)
            advantages.append(adv)
            s_t = to_tensor(s).float()
            _, _, entropy = self.policy(s_t)
            entropies.append(entropy)

        if not log_probs:
            return torch.tensor(0.0)

        log_probs_tensor = torch.stack(log_probs)
        advantages_tensor = torch.stack(advantages)
        entropies_tensor = torch.stack(entropies)
        policy_loss = -torch.mean(log_probs_tensor * advantages_tensor.detach())
        if self.entropy_coef > 0:
            entropy_loss = -self.entropy_coef * torch.mean(entropies_tensor)
            policy_loss = policy_loss + entropy_loss

        return policy_loss

    def _update_value(self) -> torch.Tensor:
        """TD error"""
        if not self.use_value_baseline or self.value_net is None:
            return torch.tensor(0.0)

        losses = []
        for traj in self.buffer.trajectories:
            for t in range(len(traj)):
                s = traj[t][0]
                r = traj[t][3]
                terminated = 1.0 if t == len(traj) - 1 else 0.0

                if t < len(traj) - 1:
                    s_next = traj[t + 1][0]
                    with torch.no_grad():
                        v_next = self.value_net(s_next)
                else:
                    v_next = torch.tensor([0.0])

                v = self.value_net(s)
                target = r + self.gamma * v_next * (1 - terminated)

                losses.append((v - target.detach()) ** 2)

        if not losses:
            return torch.tensor(0.0)
        return torch.mean(torch.cat(losses))

    def evaluate(self, state: np.ndarray) -> int:
        return self.policy.evaluate(state)
    
```

```python
from matplotlib import pyplot as plt


class TrainingLogger:
    def __init__(self):
        self.steps = []
        self.returns = []
        self.policy_losses = []
        self.value_losses = []
        self.entropies = []
        self.eval_returns = []
        
        self.fig, self.axs = None, None
        self.lines = {}
        self._initialized = False

    def log(self, step: int, episode_return: float, policy_loss: float, 
            value_loss: float, entropy: float, eval_return: Optional[float] = None):
        self.steps.append(step)
        self.returns.append(episode_return)
        self.policy_losses.append(policy_loss)
        self.value_losses.append(value_loss)
        self.entropies.append(entropy)
        if eval_return is not None:
            self.eval_returns.append(eval_return)

    def _initialize_plot(self):
        plt.ion()
        self.fig, self.axs = plt.subplots(3, 2, figsize=(14, 10))
        self.fig.suptitle('Training Progress', fontsize=16)

        colors = {'returns': 'blue', 'policy_loss': 'red', 
                  'value_loss': 'green', 'entropy': 'purple', 'eval_returns': 'orange'}

        self.lines['returns'] = self.axs[0, 0].plot([], [], color=colors['returns'], label='Episode Return')[0]
        self.axs[0, 0].set_xlabel('Steps')
        self.axs[0, 0].set_ylabel('Return')
        self.axs[0, 0].set_title('Episode Returns')
        self.axs[0, 0].legend()
        self.axs[0, 0].grid(True, alpha=0.3)

        self.lines['policy_loss'] = self.axs[0, 1].plot([], [], color=colors['policy_loss'], label='Policy Loss')[0]
        self.axs[0, 1].set_xlabel('Steps')
        self.axs[0, 1].set_ylabel('Loss')
        self.axs[0, 1].set_title('Policy Loss')
        self.axs[0, 1].legend()
        self.axs[0, 1].grid(True, alpha=0.3)

        self.lines['value_loss'] = self.axs[1, 0].plot([], [], color=colors['value_loss'], label='Value Loss')[0]
        self.axs[1, 0].set_xlabel('Steps')
        self.axs[1, 0].set_ylabel('Loss')
        self.axs[1, 0].set_title('Value Loss (if applicable)')
        self.axs[1, 0].legend()
        self.axs[1, 0].grid(True, alpha=0.3)

        self.lines['entropy'] = self.axs[1, 1].plot([], [], color=colors['entropy'], label='Entropy')[0]
        self.axs[1, 1].set_xlabel('Steps')
        self.axs[1, 1].set_ylabel('Value')
        self.axs[1, 1].set_title('Entropy & Regularization')
        self.axs[1, 1].legend()
        self.axs[1, 1].grid(True, alpha=0.3)

        self.lines['eval_returns'] = self.axs[2, 0].plot([], [], color=colors['eval_returns'], label='Eval Returns')[0]
        self.axs[2, 0].set_xlabel('Steps')
        self.axs[2, 0].set_ylabel('Return')
        self.axs[2, 0].set_title('Evaluation Returns')
        self.axs[2, 0].legend()
        self.axs[2, 0].grid(True, alpha=0.3)


        plt.tight_layout()
        self._initialized = True

    def update_plot(self):
        if not self._initialized:
            self._initialize_plot()

        if len(self.steps) == 0:
            return

        self.lines['returns'].set_data(self.steps, self.returns)
        self.lines['policy_loss'].set_data(self.steps, self.policy_losses)
        self.lines['value_loss'].set_data(self.steps, self.value_losses)
        self.lines['entropy'].set_data(self.steps, self.entropies)
        self.lines['eval_returns'].set_data(self.steps[:len(self.eval_returns)], self.eval_returns)

        for ax in self.axs.flat:
            ax.relim()
            ax.autoscale_view()

        self.fig.canvas.draw()
        self.fig.canvas.flush_events()
        plt.pause(0.01)

    def save_plot(self, filepath: str = 'training_progress.png'):
        if self._initialized:
            self.fig.savefig(filepath, dpi=150, bbox_inches='tight')
            print(f"Plot saved to {filepath}")

    def get_metrics(self) -> dict:
        return {
            'steps': np.array(self.steps),
            'returns': np.array(self.returns),
            'policy_losses': np.array(self.policy_losses),
            'value_losses': np.array(self.value_losses),
            'entropies': np.array(self.entropies)
        }
```

```python
import time
from tqdm import tqdm
def run_training_loop(
    env_name: str = "CartPole-v1",
    total_max_steps: int = 100_000,
    eval_interval: int = 2000,
    smooth_ret_window: int = 100,
    hidden_dims: List[int] = None,
    lr: float = 3e-4,
    gamma: float = 0.995,
    success_ret: float = 475.0,
    baseline: str = "rloo",
    entropy_coef: float = 0.01,
    entropy_schedule: bool = False,
    verbose: bool = True,
):
    env = gym.make(env_name)
    episode_return_history = deque(maxlen=smooth_ret_window)

    hidden_dims = hidden_dims or [64, 64]

    agent = PolicyAgent(
        state_dim=env.observation_space.shape[0],
        action_dim=env.action_space.n,
        hidden_dims=hidden_dims,
        lr=lr,
        gamma=gamma,
        baseline=baseline,
        entropy_coef=entropy_coef,
    )

    logger = TrainingLogger()

    s, _ = env.reset()
    done = False
    episode_return = 0.0
    is_eval = False
    plan_eval = False
    start_time = time.time()
    policy_loss, value_loss = 0.0, 0.0
    avg_return = 0.0
    pbar = tqdm(range(1, total_max_steps + 1), position=0, leave=True)
    for global_step in pbar:

        action, log_prob, entropy = agent.act(s)
        s_next, r, terminated, truncated, _ = env.step(action)
        episode_return += r
        done = terminated or truncated
        agent.store_step(s, action, r, log_prob, done)
        current_entropy_coef = entropy_coef
        if entropy_schedule:
            current_entropy_coef = max(0.0, entropy_coef * (1 - global_step / total_max_steps))
        losses = agent.update(entropy_coef=current_entropy_coef)
        if losses is not None:
            policy_loss, value_loss = losses

        if global_step % eval_interval == 0:
            plan_eval = True

        s = s_next
        if done:
            if is_eval:

                episode_return_history.append(episode_return)
                avg_return = np.mean(episode_return_history)
                avg_entropy = entropy.mean().item()

                s, _ = env.reset()
                eval_return = 0
                done = False
                while not done:
                    action = agent.evaluate(s)
                    s, r, terminated, truncated, _ = env.step(action)
                    eval_return += r
                    done = terminated or truncated

                logger.log(
                    step=global_step,
                    episode_return=avg_return,
                    policy_loss=policy_loss,
                    value_loss=value_loss,
                    entropy=avg_entropy,
                    eval_return=eval_return
                )

                if verbose:
                    elapsed = time.time() - start_time
                    pbar.set_description(f"step={global_step} | Return={avg_return:>7.3f} | "
                                     f"Policy L.={policy_loss:>8.5f} | Value L.={value_loss:>8.5f} | "
                                     f"Entropy={avg_entropy:>8.5f} | Time={elapsed:>6.1f}s | Eval R.={eval_return:>7.3f}")
                    
                    
                if avg_return >= success_ret:
                    print("Решено!")
                    break
            s, _ = env.reset()
            done = False
            episode_return = 0.0
            is_eval = plan_eval
            plan_eval = False

    env.close()
    return agent, logger
```

```python
import optuna

def objective(trial):
    
    baseline = trial.suggest_categorical("baseline", ("none", "value", "rloo"))
    lr = trial.suggest_float("lr", 1e-4, 1e-2, log=True)
    entropy_coef = trial.suggest_float("entropy_coef", 1e-4, 1e-1, log=True)
    hidden_dims = trial.suggest_categorical("hidden_dims", ([32, 32], [64, 64], [128, 64]))
    gamma = trial.suggest_float("gamma", 0.95, 0.999, step=0.002)

    _, logger = run_training_loop(
        total_max_steps=20_000,
        eval_interval=500,
        baseline=baseline,
        lr=lr,
        entropy_coef=entropy_coef,
        hidden_dims=hidden_dims,
        gamma=gamma,
        verbose=False
    )
    
    if len(logger.returns) >= 5:
        return np.mean(logger.returns[-10:])
    return 0.0
    


study = optuna.create_study(
    direction="maximize", 
    pruner=optuna.pruners.MedianPruner(n_startup_trials=5)
)
study.optimize(objective, n_trials=20, timeout=3600, show_progress_bar=True)

print(f"Best Score: {study.best_value:.3f}")
print(f"Best Params: {study.best_params}")
```

```python
def collect_trajectories(
    agent : PolicyAgent,
    env_name: str = "CartPole-v1",
    n_episodes: int = 100,
    min_reward: float = 450.0,
    abs_bound: float = 0.2,
):
    env = gym.make(env_name)
    
    states = []
    actions = []
    count_episodes = 0
    skipped_episodes = 0
    
    pbar = tqdm(total=n_episodes, desc="Generating trajectories")
    
    while count_episodes < n_episodes:
        s, _ = env.reset(options={"low": -abs_bound, "high": abs_bound})
        
        episode_states = []
        episode_actions = []
        episode_return = 0.0
        done = False
        
        while not done:
            action = agent.evaluate(s)
            
            episode_states.append(s.copy())
            episode_actions.append(action)
            
            s_next, r, terminated, truncated, _ = env.step(action)
            episode_return += r
            done = terminated or truncated
            s = s_next
        
        if episode_return >= min_reward:
            states.extend(episode_states)
            actions.extend(episode_actions)
            count_episodes += 1
            pbar.set_postfix({"reward": f"{episode_return:.1f}", "count": count_episodes})
        else:
            skipped_episodes += 1
        
        pbar.update(1)
    
    env.close()
    
    return np.array(states, dtype=np.float32), np.array(actions, dtype=np.int64)
```

```python
class BCAgent:
    def __init__(self, s_dim, a_dim, h_dims, lr=1e-3):
        self.policy = PolicyNetwork(s_dim, a_dim, h_dims)
        self.opt = torch.optim.Adam(self.policy.parameters(), lr=lr)
        self.loss_fn = nn.CrossEntropyLoss()

    def train_step(self, states, actions):
        self.opt.zero_grad()
        loss = 0
        for i, state in enumerate(states):
            logits = self.policy.logits(state)
            loss += self.loss_fn(torch.FloatTensor(logits), actions[i])
        loss.backward()
        self.opt.step()
        return loss.item()

    def evaluate(self, s):
        return self.policy.evaluate(s)
```

```python

def train_bc(states, actions, s_dim, a_dim, h_dims=[64, 64], epochs=50, bs=64):
    agent = BCAgent(s_dim, a_dim, h_dims)
    states, actions = torch.tensor(states), torch.tensor(actions)
    for _ in tqdm(range(epochs)):
        idx = torch.randperm(len(states))
        states, actions = states[idx], actions[idx]
        for i in range(0, len(states), bs):
            agent.train_step(states[i:i+bs], actions[i:i+bs])
    return agent

env = gym.make("CartPole-v1")
s_dim, a_dim = env.observation_space.shape[0], env.action_space.n
env.close()

bc_agent = train_bc(trajs[0], trajs[1], s_dim, a_dim)
```

```python
def eval_bc(bc_agent, abs_bound, episodes=200):
    env = gym.make("CartPole-v1")
    bc_rets = []
    for _ in range(episodes):
        agent, rets = bc_agent, bc_rets
        s, _ = env.reset(options={'low': -abs_bound, 'high': abs_bound})
        
        ret, done = 0.0, False
        while not done:
            a = agent.evaluate(s)
            s, r, term, trunc, _ = env.step(a)
            ret += r
            done = term or trunc
        rets.append(ret)
    env.close()
    return np.mean(bc_rets), np.std(bc_rets) / np.sqrt(episodes)


print(f"Standard: BC={eval_bc(bc_agent, 0.05)[0]:.1f}")
for p in [0.1, 0.18, 0.20, 0.22, 0.3]:
    bc_r = eval_bc(bc_agent, p)
    print(f"Perturb {p}: BC={bc_r[0]:.1f} BC_std={bc_r[1]:.1f}")
```
